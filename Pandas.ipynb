```python
# string based data
df["Gender"] = "Female"
df.iloc[0:10, 6] = "Male"
df.head()
```

```python
def f(s):
    if s == "Male":
        return 0
    else:
        return 1

df["Sex"] = df.Gender.apply(f)
del df["Gender"]
df.head()
```
